```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
```

```python
len(tokenizer.word_index)
```

```python
tokenizer.texts_to_sequences(text.split('\n'))
```

```python
input_sequences=[]
for sentence in text.split('\n'):
    tokenized_sentences=tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentences)):
        input_sequences.append(tokenized_sentences[:i+1])
```

```python
input_sequences
```

```python
len(input_sequences)
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
X=padded_input_sequences[:,:-1]
y=padded_input_sequences[:,-1]
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=283)
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
model.add(Embedding(283,100,input_length=56,input_shape=(max_len,)))
model.add(LSTM(150,return_sequences=True))
model.add(LSTM(150))
model.add(Dense(283,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=50)
```

```python
import time

def Next_WordPredictor(text_1):
    for i in range(10):
        token_text=tokenizer.texts_to_sequences([text_1])[0]
        padded_token_text=pad_sequences([token_text],maxlen=56,padding='pre')
        pos=np.argmax(model.predict(padded_token_text))
        for word,index in tokenizer.word_index.items():
            if index==pos:
                text_1=text_1+" "+word
                print(text_1)
                time.sleep(2)
```

```python
text=input()
Next_WordPredictor(text)
```
